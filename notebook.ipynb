```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/self_reported_personal_data_breach_cases.csv")
df[df.decision == "Regulatory action taken"]
```

```python
offenders = df.submitted_about_account.value_counts()
offenders = offenders[offenders > 15]
offenders.plot(kind="barh", figsize=(5, 10), title="Most reported offenders in self-reported personal data breach cases")
plt.grid(axis="x")
```
